```python
limit = df.w.describe()["mean"]
limit = .9
limit2=4
```

```python
# df= df.head(n=10000)
df = df[df['w']< limit]
df = df[  (df['x']<limit2)  & (df['y'] < limit2)  & (df['z']<limit2)  
        & (df['x']> -limit2)  & (df['y'] > -limit2)  & (df['z'] > -limit2) ]
# df= df [  df['z']> 0 ]
```

```python
from mayavi import mlab
```

```python
#Kudos to the priginal code
#https://stackoverflow.com/questions/12423601/simplest-way-to-plot-3d-surface-given-3d-points/25586869#25586869

X = df['x'].to_numpy()
Y = df['y'].to_numpy()
Z = df['z'].to_numpy()
W = df['w'].to_numpy()

# Define the points in 3D space
# including color code based on Z coordinate.
pts = mlab.points3d(X, Y, Z)

# Triangulate based on X, Y with Delaunay 2D algorithm.
# Save resulting triangulation.
mesh = mlab.pipeline.delaunay2d(pts)

# Remove the point representation from the plot
pts.remove()
# pts = mlab.points3d(df_my_points['x'],df_my_points['y'],df_my_points['z'],df_my_points['code'])
pts = mlab.points3d(df_my_points_blue['x'],
                    df_my_points_blue['y'],
                    df_my_points_blue['z'],color=(0.0, 0.65, 0.255)
                    ,scale_factor=.1
                   )

pts = mlab.points3d(df_my_points_red['x'],
                    df_my_points_red['y'],
                    df_my_points_red['z'],color=(0.255, 0.237, 0.00)
                    ,scale_factor=.1
                   )
# Draw a surface based on the triangulation
surf = mlab.pipeline.surface(mesh,opacity=0.50)

# Simple plot.
mlab.xlabel("x")
mlab.ylabel("y")
mlab.zlabel("z")
mlab.show()
```
